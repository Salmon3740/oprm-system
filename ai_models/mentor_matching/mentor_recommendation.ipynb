```python
import pandas as pd
df = pd.read_csv("professors_dataset.csv")

df.head()
```

```python
df["expertise_domains"] = df["expertise_domains"].apply(lambda x: x.split(","))
df.head()
```

```python
def recommend_mentors(predicted_domains, top_n=5):

    recommendations = []

    for index, row in df.iterrows():

        professor_domains = [d.strip() for d in row["expertise_domains"]]

        # Calculate matching score
        score = len(set(predicted_domains) & set(professor_domains))

        if score > 0 and row["availability"] == "Available":
            recommendations.append((row["name"], professor_domains, score))

    recommendations.sort(key=lambda x: x[2], reverse=True)

    return recommendations[:top_n]
```

```python
domains = ["Machine Learning", "Data Science"]

recommend_mentors(domains)
```
